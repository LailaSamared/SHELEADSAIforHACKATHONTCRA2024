```python
#check for missing value
missing_values = hdl.isnull().sum()

missing_values
```

```python
#convert datatype
cols_to_convert = ['Gender Inequality Index', 'Rank', 'Maternal mortality ratio',
                   'Adolescent birth rate', 'F_secondary_educ', 'M_secondary_educ',
                   'F_Labour_force', 'M_Labour_force','Share of seats in parliament (% held by women)']

for col in cols_to_convert:
    hdl[col] = pd.to_numeric(hdl[col], errors='coerce')
hdl.info()
```

```python
#check the missing value again
hdl.isnull().sum()
```

```python
#fill the missing value
cols_to_fill = ['Gender Inequality Index', 'Rank', 'Maternal mortality ratio',
                'F_secondary_educ', 'M_secondary_educ', 'F_Labour_force', 'M_Labour_force','Share of seats in parliament (% held by women)']

for col in cols_to_fill:
    hdl[col] = hdl[col].fillna(hdl[col].median())
hdl.isnull().sum()
```
